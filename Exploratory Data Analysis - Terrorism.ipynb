## Perform Exploratory Data Analysis on the dataset 'Global Terrorism'

```python
# Importing all the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the given dataset as data

data = pd.read_csv("globalterrorism.csv", encoding='latin-1')
```

```python
# Checking the column names in the given data

data.columns
```

```python
# Changing the column names to more meaningful and descriptive column names

data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                  'region_txt':'Region','attacktype1_txt':'Attack_type','target1':'Target','nkill':'Killed',
                  'nwound':'wounded', 'gname':'Group','targettype1_txt':'Target_type','weaptype1_txt':'Weapon_Type'},
          inplace=True)
```

```python
# Rechecking the column names of the given data

data.columns
```

```python
# Checking for missing Values in the given dataset

missing_values = data.isnull().sum()
missing_values
```

```python
# Checking the total count of values

np.prod(data.shape)
```

```python
# Checking the percentage of missing values

missing_per = missing_values.sum()/24528285
missing_per*100
```

Here we can notice that the data comprises of a huge number of missing values

Hence, lets now create a new dataset by dropping the columns with missing values

```python
# Creating a dataset new_data after dropping the missing values

new_data = data.dropna(axis = 1)
```

```python
# Creating a new dataset with only those columns necessary for further analysis

data_final = data[['Year','Month','Day','Country','State','Region','city','latitude','longitude','Attack_type','Killed',
      'wounded','Target','summary','Group','Weapon_Type','motive']]
```

```python
# Replacing the missing values with integer values

data_final['wounded'] = data_final['wounded'].fillna(0).astype(int)
data_final['Killed'] = data_final['Killed'].fillna(0).astype(int)
```

### Exploratory Data Analysis

```python
# Creating a countplot to visualize the countries that have the highest number of terrorist attacks

plt.subplots(figsize=(20,10))
sns.countplot('Country', data = data_final)
plt.xticks(rotation=90)
plt.title("Terrorist attacks by country")
plt.show()
```

Here we can notice that it is nearly impossible to visualize this data, thus let us just find the name of the country with the most number of terrorist attacks

```python
# Country with most terrorist attacks

country = data_final['Country'].value_counts().idxmax()
print('The country with the most terrorist attacks is', country)
```

```python
# City with most terrorist attacks

city = data_final['city'].value_counts().index[1]
print('The city with the most terrorist attacks is', city)
```

```python
# Region with most terrorist attacks

region = data_final['Region'].value_counts().idxmax()
print('The country with the most terrorist attacks is', region)
```

```python
# Selecting the top 10 countries with the most number of people killed due to terrorist activities

plt.subplots(figsize=(13,6))
data_final.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(10).plot(kind = 'bar', color = 'crimson')
plt.xticks(rotation = 90)
plt.ylabel('Killed')
plt.title("Top 10 countries with the most number of people killed due to terrorist activities)")
plt.show()
```

<b>Iraq</b> has the highest death rate of over 70000 deaths caused due to terrorist activities.

```python
# Selecting the top 10 countries with the most number of people wounded due to terrorist activities

plt.subplots(figsize=(13,6))
data_final.groupby(['Country'])['wounded'].sum().sort_values(ascending=False).head(10).plot(kind = 'bar', color = 'midnightblue')
plt.xticks(rotation = 90)
plt.ylabel('Wounded')
plt.title("Top 10 countries with the most number of people wounded due to terrorist activities)")
plt.show()
```

<b>Iraq</b> has the highest wounded people rate of 130000 deaths caused due to terrorist activities.

```python
# The number of terrorist attacks that have taken place over the past years

plt.subplots(figsize = (20,10))
sns.countplot('Year', data = data_final, palette = 'deep')
plt.xticks(rotation = 90)
plt.title("The number of terrorist attacks that have taken place over the past years")
plt.show()
```

Here we can notice that the year <b>2014</b> has been the worst year considering that more than 1600 terrorist attacks had taken place in 2014.

```python
# Ratio of people killed in specific regions

plt.subplots(figsize=(10,5))
sns.barplot(x = 'Region', y = 'Killed', data = data_final)
plt.xticks(rotation=90)
plt.ylabel("Killed")
plt.title("Ratio of people killed in specific regions")
plt.show()
```

<b>Sub-Saharan Africa</b> has the highest ratio of people being killed by terrorists.

```python
# Ratio of people wounded in specific regions

plt.subplots(figsize=(10,5))
sns.barplot(x = 'Region', y = 'wounded', data = data_final)
plt.xticks(rotation=90)
plt.ylabel("Wounded")
plt.title("Ratio of people wounded in specific regions")
plt.show()
```

<b>East Asia</b> has the highest number of wounded people caused by terrorist activities.

```python
# Weapon used for the attack

plt.subplots(figsize=(10,10))
sns.countplot('Weapon_Type', data = data_final, order = data_final['Weapon_Type'].value_counts().index)
plt.xticks(rotation = 90)
plt.title("Weapon used for the attack")
plt.show()
```

<b>Explosives and Firearms</b> are the most used weapon by terrorists

```python
# Method of Attack
plt.subplots(figsize=(10,10))
sns.countplot('Attack_type', data = data_final, order = data_final['Attack_type'].value_counts().index)
plt.xticks(rotation = 90)
plt.title("Method of Attack")
plt.show()
```

<b>Bombing/Explosion</b> is the most used Attack type by terrorists.
